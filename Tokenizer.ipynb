```python
file_path = './dataset/the-verdict.txt'
with open(file_path, "r",encoding='utf-8') as f:
    raw_text = f.read()
print(raw_text[:99])
print(len(raw_text))
```

## Tokenizer
使用正则表达式划分单词

### Create vocabulary

```python
# 分割文本中的单词及标点
words = re.split(r'([,.?_!"()\']|--|\s)',raw_text)
# 标点被识别为空格，去除空格
words =  [word.strip() for word in words if word.strip()]
# 将字符转换为set类型，并进行排序
all_words = sorted(set(words))
print(all_words[:50])
vocabulary = {token: index for index, token in enumerate(all_words)}

for key, value in vocabulary.items():
    if value > 50:
        break
```

```python
class Tokenizerv1:
    def __init__(self, vocabulary):
        self.vocab = vocabulary
        self.vocab_len = len(vocabulary)
        # set.item每次迭代返回的是键值对
        self.int_to_str = {value: key for key, value in self.vocab.items()}
    def encoder(self, text):
        # 将输入字符进行分割
        tokens = re.split(r'([,.?_!"()\']|--|\s)', text)
        # 去除空格字符
        tokens =  [word.strip() for token in tokens if token.strip()]
        ids = [self.vocab[token]  for token in tokens]
        return ids
    def decoder(self, ids):
        text = "".join([self.int_to_str[id]  for id in ids])
        return text
```

```python
tokenizer = Tokenizerv1(vocabulary)
```
